```python
from pymongo import MongoClient
import pymongo
```

```python
# establish connection to the mongodb
client = MongoClient("mongodb://127.0.0.2:27017")   

print(client.list_database_names())
```

```python
housing = client["housing"]
market = housing["market"]
```

```python
for row in market.find().limit(1):
    print(row)
```

```python

market.find_one({"median_sale_price": 175501})
```

```python
#Search for a specific city
#test case: Eagle Lake
input_Q2 = input("enter name of city ")
```

```python
result_Q2 = market.find({'city': input_Q2},{'city': 1, 'state': 1, 'median_sale_price':1})
#only display city, state, median_sale_price
```

```python
for i in result_Q2:
    print(i)
```

```python
#Search for a specific time frame
```

```python
year = input("Enter year with format xxxx")
month = input("Enter month with format xx")
time = "{}-{}-01".format(year, month)
time
```

```python
result_Q3 = market.aggregate(
    [{"$match" : {"period_begin": {"$in" :[time]}}}
    ])
```

```python
for i in result_Q3:
    print(i)
```

```python
#Show the average median sale price for a specific region
```

```python
#test case: 'Eagle Lake, WI'
city = input('Enter the city ')
state_code = input('Enter state code ')
```

```python
region = "{}, {}".format(city,state_code)
region
```

```python
result_Q7 = market.aggregate(
    [{"$match" : {"region" : region }},
    { '$group':{ '_id' : '$region', 'Average median sale price' : { '$avg': '$median_sale_price' }}}
    ])
```

```python
for i in result_Q7:
    print(i)
```

```python
#Compare the average median sale prices from multiple regions
```

```python
#test case: 'Eagle Lake, WI','Double Oak, TX'
print("Compare avarage median sale price bettwen 2 regions \n")
city1 = input('Enter the city of the first region ')
state_code1 = input('Enter state code of the first region ')
city2 = input('Enter the city of the seconf region ')
state_code2 = input('Enter state code of the second region ')
```

```python
region1 = "{}, {}".format(city1,state_code1)
region2 = "{}, {}".format(city2,state_code2)
region1, region2
```

```python
result_Q9 = market.aggregate(
    [{"$match" : {"region": {"$in" : [region1, region2 ]}}},
    { '$group':{ '_id' : '$region', 'sum' : { '$avg': '$median_sale_price' }}}
    ])
```

```python
for i in result_Q9:
    print(i)
```
